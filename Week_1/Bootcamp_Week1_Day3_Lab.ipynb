```python
def daily_summary(matrix):
    output_list = []
    
    for i in range(len(matrix)):
        day_row = matrix[i]
        
        stats = {
            'day': i + 1,
            'min': day_row.min(),
            'max': day_row.max(),
            'mean': day_row.mean()
        }
        output_list.append(stats)
        
    return output_list

temp_summaries = daily_summary(temp_matrix)
print(temp_summaries[0])
```
